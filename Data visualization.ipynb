```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df = pd.read_csv('cleaned data.csv')
df.describe()
```

```python
df.corr()
```

## Horse Power Vs Price

```python
sns.regplot(x='Power', y='Price',data = df)
plt.ylim(0,)
```

From the plot we can see that there is a positive direct correlation between <b>Horse Power</b> and <b> Price</b>.
So Horse Power is good predictor of price. 

## Engine Vs Price

```python
sns.regplot(x='Engine', y='Price',data = df)
```

For Engine size also we can see a positive correaltion between <b> Engine Size</b> and <b> Price</b>.

## Kilometers Driven Vs Price

```python
df = df[df['Kilometers_Driven'] < 6000000]
```

```python
sns.regplot(x='Kilometers_Driven', y='Price', data = df)
```

From the regression plot we can see that Price of Car is decreasing as Kilometer Driven is increasing. So we can assume<br> that odometer reading is negatively correlated to Price and its is good predictor of price.
